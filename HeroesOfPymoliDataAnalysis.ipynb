```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)
csv_path = "~/Desktop/HeroesofPymoli/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(csv_path)
```

Player Count

Display the total number of players

```python
purchase_data.head()
```

```python
#streamline data by gender age and SN

player_count = purchase_data.loc[:, ["Gender", "SN", "Age"]]
player_count.head()
```

```python
#sn has duplicate players, drop duplicates to determine accurate number

player_count = player_count.drop_duplicates()
cleanplayers = player_count.count()[0]
cleanplayers
```

```python
#display data frame of clea players 

pd.DataFrame({"Total players": [cleanplayers]})
```

Purchasing Analysis (Total)

Run basic calculations to obtain number of unique items, average price, etc.

Create a summary data frame to hold the results

Optional: give the displayed data cleaner formatting

Display the summary data frame

```python
# Collect a list of the unique values
# Calculate averages for price using the .mean() method
# Calculate total for purchase using the .count() method
# Calculate total for revenue using the .sum() method
uniqueitems = len(purchase_data["Item ID"].unique())
averageprice = purchase_data["Price"].mean()
totalpurchases = purchase_data["Price"].count()
totalrevenue = purchase_data["Price"].sum()

# Convert a single dictionary containing lists into a dataframe
# Rename the headers to be more explanatory
Purchasing_analysis= pd.DataFrame({"Number of Unique Items": [uniqueitems],
                             "Average Purchase Price": [averageprice],
                             "Total Number of Purchases": [totalpurchases],
                             "Total Revenue": [totalrevenue]})

Purchasing_analysis.head()
```

Gender Demographics

Percentage and Count of Male Players

Percentage and Count of Female Players

Percentage and Count of Other / Non-Disclosed

```python
# sort data by gender
# use math to create percentage
genderdemographics = player_count["Gender"].value_counts()
percentage = genderdemographics/ cleanplayers * 100

# Rename the headers to be more explanatory
genderdemographics = pd.DataFrame({"Total Count": genderdemographics,
                                    "Percentage of Players": percentage})

genderdemographics = genderdemographics.round()

genderdemographics.head()
```

Purchasing Analysis (Gender)

Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

Create a summary data frame to hold the results

Optional: give the displayed data cleaner formatting

Display the summary data frame

```python
# Using GroupBy in order to separate the data into fields according to "gender" values
# Calculate total for purchase using the .count() method
# Calculate averages for price using the .mean() method
# Calculate total for revenue using the .sum() method
# use math to create percentage

purchase_count = purchase_data.groupby(["Gender"]).count()["Price"]
avg_purchase_price = purchase_data.groupby(["Gender"]).mean()["Price"]
total_purchase_value = purchase_data.groupby(["Gender"]).sum()["Price"]
Avg_total_per_person = total_purchase_value / player_count["Gender"].value_counts()

# create datafram
purchasing_analysis_gender = pd.DataFrame({"Purchase Count": purchase_count,
                                         "Average Purchase Price": avg_purchase_price,
                                         "Total Purchase Value": total_purchase_value,
                                         "Avg Total Purchase per Person": Avg_total_per_person})


purchasing_analysis_gender = purchasing_analysis_gender[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]

#round the results to 2 decimals
purchasing_analysis_gender = purchasing_analysis_gender.round(2)
purchasing_analysis_gender.head()
```

Age Demographics

Establish bins for ages

Categorize the existing players using the age bins. Hint: use pd.cut()

Calculate the numbers and percentages by age group

Create a summary data frame to hold the results

Optional: round the percentage column to two decimal points

Display Age Demographics Table

```python
#create bins
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]

#create names for bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#looking by age  using pd.cut 
#Use cut when you need to segment and sort data values into bins
age_demographics = player_count.loc[:,["Age"]]
age_demographics["Group Names"] = pd.cut(age_demographics["Age"], bins, labels=group_names)

#create count per bin by number of players per age count
#percentage of players within age bins
total_count = age_demographics["Group Names"].value_counts()
percentage_of_players = total_count / cleanplayers * 100

#create dataframe
age_demographics = pd.DataFrame({"Total Count": total_count, "Percentage of Players": percentage_of_players})

#display index round to two decimal places
age_demographics = age_demographics.sort_index()

age_demographics.round(2)
```

Purchasing Analysis (Age)

Bin the purchase_data data frame by age

Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below

Create a summary data frame to hold the results

Optional: give the displayed data cleaner formatting
    
Display the summary data frame

```python
#create bins
bins2 = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]

#create names for bins
group_names2 = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#looking by age  using pd.cut 
#Use cut when you need to segment and sort data values into bins
pd.cut(purchase_data["Age"], bins2, labels = group_names2)

purchase_data["Group Names2"] = pd.cut(purchase_data["Age"],bins2,labels = group_names2)

# Using GroupBy in order to separate the data into fields according to "gender" values
# Calculate total for purchase using the .count() method
# Calculate averages for price using the .mean() method
# Calculate total for revenue using the .sum() method
# use math to create percentage

purchase_count = purchase_data.groupby(["Group Names2"]).count()["Age"]
avg_purchase_price = purchase_data.groupby(["Group Names2"]).mean()["Price"]
total_purchase_value= purchase_data.groupby(["Group Names2"]).sum()["Price"]
Avg_total_per_person = total_purchase_value /purchase_count


#create dataframe
purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                         "Average Purchase Price": avg_purchase_price,
                                         "Total Purchase Value": total_purchase_value,
                                         "Avg Total Purchase per Person": Avg_total_per_person})


purchasing_analysis = purchasing_analysis[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]

#display index round to two decimal places
purchasing_analysis = purchasing_analysis.sort_index()
purchasing_analysis.round(2)
```

Top Spenders

Run basic calculations to obtain the results in the table below

Create a summary data frame to hold the results

Sort the total purchase value column in descending order

Optional: give the displayed data cleaner formatting

Display a preview of the summary data frame

```python
# Using GroupBy in order to separate the data into fields according to "SN" values to see top spenders
# Calculate total for purchase using the .count() method
# Calculate averages for price using the .mean() method
# Calculate total for value using the .sum() method

top_spenders=purchase_data.groupby(["SN"]).count()["Price"]
average=purchase_data.groupby(["SN"]).mean()["Price"]
total=purchase_data.groupby(["SN"]).sum()["Price"]

#create dataframe
summary=pd.DataFrame({"Purchase Count":top_spenders,
                          "Average Purchase":average,
                          "Total Purchase Value":total})

# It is possible to sort based upon multiple columns
summary.sort_values("Total Purchase Value",ascending=False).round(2).head()
```

Most Popular Items

Retrieve the Item ID, Item Name, and Item Price columns

Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value

Create a summary data frame to hold the results

Sort the purchase count column in descending order

Optional: give the displayed data cleaner formatting

Display a preview of the summary data frame

```python
# Using GroupBy in order to separate the data into fields according to "item id" and "item name" 
# calculate using count, mean and sum
# Rename columns for readability
purchase_count2= purchase_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
item_price = purchase_data.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Average Purchase Price")
total_purchase_value = purchase_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")


#create dataframe
summary = pd.DataFrame({"Purchase Count":purchase_count2,
                                   "Item Price":item_price,
                                   "Total Purchase Value":total_purchase_value,})

#Sort Descending by pruchase count
most_popular_items = summary.sort_values("Purchase Count", ascending=False)
most_popular_items.head()
```
